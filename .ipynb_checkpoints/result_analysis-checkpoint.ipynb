```python
import pandas as pd
import networkx as nx

def create_network_graph(path, data_name, feature_file, full_network_file, label_name, ratio):
    # Construct file paths
    feature_path = f"{path}/{data_name}/{feature_file}"
    network_file_path = f"{path}/{data_name}/{full_network_file}"
    
    # Read feature data
    df_features = pd.read_csv(feature_path)
    feature_names = list(df_features.drop(columns=[label_name]).columns)
    
    # Read the full network data
    df = pd.read_csv(network_file_path)
    df = df.sort_values(by='IG', ascending=False)
    
    # Filter based on the given ratio
    if int(df.shape[0] * ratio) < 1:
        df_filter = df[:1]
    else:
        df_filter = df[:int(df.shape[0] * ratio)]
    
    # Drop unnecessary columns and save the filtered network
    df_filter = df_filter.drop(columns=['IG', 'MutualInfo'])
    optimal_network_file = f"{path}/{data_name}/comprehensive_network_optimal.csv"
    df_filter.to_csv(optimal_network_file, index=False)
    
    # Read the filtered network file to extract edges
    with open(optimal_network_file, 'r') as file:
        edges = [line.strip().split(',')[0:2] for line in file]
    
    # Create and populate the graph
    G = nx.Graph()
    G.add_nodes_from(feature_names)
    G.add_edges_from(edges[1:])  # Skip header
    
    return G

```

```python
import matplotlib.pyplot as plt
import networkx as nx
from adjustText import adjust_text
import matplotlib.cm as cm
import math

def plot_network_graphs(G_dict,save_path):
    """
    Plots multiple network graphs in a dynamic grid of subplots.
    
    Parameters:
    G_dict (dict of {str: nx.Graph}): Dictionary with titles as keys and network graphs as values.
    """
    num_graphs = len(G_dict)
    
    # Calculate grid size
    cols = min(3, num_graphs)  # Up to 3 columns
    rows = math.ceil(num_graphs / cols)  # Number of rows needed
    
    fig, axes = plt.subplots(rows, cols, figsize=(6 * cols, 6 * rows))  # Create subplots grid
    axes = axes.flatten() if num_graphs > 1 else [axes]  # Flatten axes array, handle single plot case
    
    for i, (title, G) in enumerate(G_dict.items()):
        ax = axes[i]
        
        # Identify isolated and connected nodes
        isolated_nodes = [node for node, degree in G.degree() if degree == 0]
        connected_nodes = [node for node, degree in G.degree() if degree > 0]

        # Compute positions for connected nodes
        pos = nx.spring_layout(G, seed=42, k=0.3, scale=2)  # Tighter layout

        # Define a specific area for isolated nodes (close to the main graph)
        if isolated_nodes:
            isolated_x = min(pos[node][0] for node in connected_nodes) - 0.2
            isolated_y = sum(pos[node][1] for node in connected_nodes) / len(connected_nodes) - 0.25
            
            # Position isolated nodes in a clustered area close to the main network
            for j, node in enumerate(isolated_nodes):
                pos[node] = (isolated_x, isolated_y + j * 0.2)  # Adjust the vertical spacing

        # Get node degrees to use as color mapping
        degrees = dict(G.degree())

        # Choose a list of distinct colors from the 'tab10' colormap
        cmap = plt.cm.get_cmap('tab10', len(connected_nodes))  # Color map for connected nodes
        isolated_cmap = 'lightgray'  # Color for isolated nodes

        # Assign colors to nodes
        node_colors = []
        for node in G.nodes():
            if node in isolated_nodes:
                node_colors.append(isolated_cmap)  # Assign different color to isolated nodes
            else:
                node_colors.append(cmap(degrees[node] % cmap.N))  # Use cmap for connected nodes

        # Draw nodes with custom colors
        nx.draw_networkx_nodes(
            G, pos, ax=ax, node_size=600, cmap=cmap,
            node_color=node_colors, alpha=0.9, edgecolors='white'
        )

        # Draw edges with custom styles
        nx.draw_networkx_edges(
            G, pos, ax=ax, width=2.0, alpha=0.5, edge_color='gray', style='dashed'
        )

        # Draw labels with enhanced font properties
        label_texts = []
        for node, coords in pos.items():
            x, y = coords
            label = ax.text(x, y, s=str(node), fontsize=14, fontweight='bold',
                            color='darkblue', family='sans-serif', ha='center', va='center')
            label_texts.append(label)

        # Adjust text to prevent overlap (uncomment if needed)
        # adjust_text(label_texts, arrowprops=dict(arrowstyle='-', color='gray', lw=0.5), ax=ax)

        # Set the title for each subplot
        ax.set_title(title, fontsize=20, fontweight='bold')
        ax.axis('off')  # Hide axes

    # Remove extra subplots if the number of graphs is less than the grid
    if num_graphs < rows * cols:
        for j in range(num_graphs, rows * cols):
            fig.delaxes(axes[j])

    plt.tight_layout()
    plt.subplots_adjust(wspace=0.3, hspace=0.3)
    plt.savefig(save_path) 

    plt.show()


```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from matplotlib.ticker import MaxNLocator,MultipleLocator
import math

def plot_multiple_graphs(data_dict, sample_step=5):
    """
    Plots multiple line graphs in a dynamic grid of subplots with enhanced styling.
    
    Parameters:
    data_dict (dict): Dictionary with subplot titles as keys and lists of data as values.
    sample_step (int): Step size for sampling data points for plotting. Default is 5.
    """
    num_plots = len(data_dict)
    
    # Calculate grid size
    cols = min(3, num_plots)  # Up to 3 columns
    rows = math.ceil(num_plots / cols)  # Number of rows needed
    
    # Set global aesthetics for plots
    sns.set(style='whitegrid')  # Seaborn style for clean grid aesthetics
    
    fig, axes = plt.subplots(rows, cols, figsize=(7 * cols, 5 * rows), constrained_layout=True)  # Create subplots grid
    axes = axes.flatten() if num_plots > 1 else [axes]  # Flatten axes array, handle single plot case

    # Define a color palette
    colors = sns.color_palette("husl", num_plots)  # Husl color palette for distinct colors
    
    for i, (title, data) in enumerate(data_dict.items()):
        ax = axes[i]
        
        # Sample data every 'sample_step' points
        sampled_indices = np.arange(0, len(data), sample_step)
        sampled_data = [data[j] for j in sampled_indices]
        
        
        # Generate a smoother line with interpolation based on sampled data
        x = sampled_indices * 0.02
        x_smooth = np.linspace(x.min(), x.max(), 300) 
        y_smooth = np.interp(x_smooth, x, sampled_data)
        # print(x_smooth)
        
        # Plot the smooth line
        ax.plot(x_smooth, y_smooth, marker='', linestyle='-', color=colors[i], lw=2.5, alpha=0.85)
        
        # Scatter plot for sampled data points
        ax.scatter(x, sampled_data, color=colors[i], s=50, edgecolor='black', zorder=3)

        # Set title with bold font
        ax.set_title(title, fontsize=22, fontweight='bold')

        # Optional: Customize x and y labels
        ax.set_xlabel('Proportion of retained edges', fontsize=18, fontweight='medium')
        ax.set_ylabel('F1 Score', fontsize=18, fontweight='medium')

        # Set the limits and grid for better aesthetics
        ax.grid(True, which='both', linestyle='--', linewidth=0.6, alpha=0.7)
        ax.margins(0.05)  # Add a small margin around the plot
       # ax.xaxis.set_major_locator(MaxNLocator(integer=True))  # Ensure x-axis has integer values
     #   ax.xaxis.set_major_locator(MultipleLocator(0.1))  # Interval of 0.1 on x-axis


        # Remove top and right borders for a cleaner look
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)

    # Remove extra subplots if the number of graphs is less than the grid
    if num_plots < rows * cols:
        for j in range(num_plots, rows * cols):
            fig.delaxes(axes[j])
    plt.savefig('./results/pics/fig_F1_ratios_1.pdf',bbox_inches='tight') 

    plt.show()

```
